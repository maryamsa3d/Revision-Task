```python
data_set = pd.merge(book, borrowings, on='BookID')
data_set = pd.merge(data_set, members, on='MemberID')
```

```python
data_set.info()
```

```python
data_set["Genre"].unique()
```

```python
genre = "History"
prob_genre = (data_set["Genre"] == genre).mean()
print(f"Probability is {genre} book: {prob_genre:.2f}")
```

```python
data_set["MembershipType"].unique()
```

```python
membership = "Public"

prob_member = (data_set["MembershipType"] == membership).mean()

print(f"Probability : {membership} member: {prob_member:.2f}")
```

```python
prob_two_conditions = (
    (
        (data_set["Genre"] == "Science Fiction") &
        (data_set["MembershipType"] == "Premium")
    ).mean()
)

print(f"Probability of Science Fiction book borrowed by a Premium member: {prob_two_conditions:.2f}")
```

```python
prob_two_conditions = (
    (
        (data_set["Genre"] == "History") &
        (data_set["MembershipType"] == "Public")
    ).mean()
)
print(f"Probability : {prob_two_conditions:.2f}")
```

```python
genre_count = data_set["Genre"].value_counts()

genre_count
```

```python
print("Highest Category:")
print(genre_count.index[0], genre_count.iloc[0])

print("Lowest Category:")
print(genre_count.index[-1], genre_count.iloc[-1])
```

```python
memberco = data_set["Name"].value_counts()
print(memberco)
print("Most frequent member:")
print(memberco.index[0], memberco.iloc[0])
```

```python
publication = data_set.groupby("PublicationYear")["BorrowID"].count()
print(publication)
```

```python
authors = data_set["Author"].value_counts()

print(authors)
print("Most popular author:")
print(authors.index[0], authors.iloc[0])
```

```python
never_borrowed = book[~book["BookID"].isin(borrowings["BookID"])]
never_borrowed
```

```python
data_set["BorrowDate"] = pd.to_datetime(data_set["BorrowDate"])
data_set["ReturnDate"] = pd.to_datetime(data_set["ReturnDate"])
```

```python
data_set["BorrowDuration"] = (data_set["ReturnDate"] - data_set["BorrowDate"]).dt.days
```

```python
data_set.groupby("Genre")["BorrowDuration"].mean()
```
